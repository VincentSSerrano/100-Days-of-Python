```python
# 1. Define the tool
def greet_expert(name):
    message = f"Hello {name}, you are learning functions!"
    return message

# 2. Use the tool (This is called "Calling" the function)
# We pass the string "Student" into the 'name' label
output = greet_expert(input("Your Name: "))

print(output)
```
